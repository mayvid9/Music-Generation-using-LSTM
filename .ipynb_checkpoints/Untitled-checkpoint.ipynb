```python
from music21 import converter,instrument,note,chord
import numpy as np
import glob
import pickle
from keras.models import Sequential
from keras.layers import Dense,LSTM,Activation,Dropout
from keras.utils import np_utils
```

```python
data=[]
for files in glob.glob("demo_songs/*.mid"):
    midi=converter.parse(files)
    #print(files) 
    part=instrument.partitionByInstrument(midi) 
    #print(part.parts[0].recurse())
    notes=None
    if part:
        notes=part.parts[0].recurse()
    else:
        notes=midi.flat.notes
    for i in notes:
        if isinstance(i,note.Note):
            data.append(str(i.pitch))
        elif isinstance(i,chord.Chord):
            #print('hey')
            data.append('.'.join(str(n) for n in i.normalOrder)) 
        
    
```

```python
with open("converted_data",'wb') as file:
    pickle.dump(data,file) 
```

```python
print(data) 
```

```python
seq_len=100
pitch=sorted(set(i for i in data))
note_to_int=dict((note,i) for i,note in enumerate(pitch)) 
```

```python
note_to_int
```

```python
input=[] 
output=[] 
for i in range(len(data)-seq_len):
    inp=data[i:i+seq_len]
    out=data[i+seq_len]
    input.append([note_to_int[j] for j in inp]) 
    output.append(note_to_int[out]) 
```

```python
len(input),len(output) 
```

```python
input=np.reshape(input,(len(input),seq_len,1))
input=input/len(set(data)) 
```

```python
output=np_utils.to_categorical(output) 
```

```python
model=Sequential()
model.add(LSTM(512,input_shape=(input.shape[1],input.shape[2]),return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(len(set(data)),activation='softmax')) 
model.compile(loss='categorical_crossentropy',optimizer='rmsprop')
model.fit(input,output,epochs=200,batch_size=64) 
```
